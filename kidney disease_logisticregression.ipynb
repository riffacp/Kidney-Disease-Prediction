```python
#for str data
data['rbc']=data['rbc'].fillna(method='ffill')
data['pc']=data['pc'].fillna(method='ffill')
data['pcc']=data['pcc'].fillna(method='ffill')
data['ba']=data['ba'].fillna(method='ffill')
data['htn']=data['htn'].fillna(method='ffill')
data['dm']=data['dm'].fillna(method='ffill')
data['cad']=data['cad'].fillna(method='ffill')
data['appet']=data['appet'].fillna(method='ffill')
data['pe']=data['pe'].fillna(method='ffill')
data['ane']=data['ane'].fillna(method='ffill')
data['classification']=data['classification'].fillna(method='ffill')
```

```python
#for numerical datadata
data['age']=data['age'].fillna(data['age'].median())
data['bp']=data['bp'].fillna(data['bp'].median())
data['sg']=data['sg'].fillna(data['sg'].median())
data['al']=data['al'].fillna(data['al'].median())
data['su']=data['su'].fillna(data['su'].median())
data['bgr']=data['bgr'].fillna(data['bgr'].median())
data['bu']=data['bu'].fillna(data['bu'].median())
data['sc']=data['sc'].fillna(data['sc'].median())
data['sod']=data['sod'].fillna(data['sod'].median())
data['pot']=data['pot'].fillna(data['pot'].median())
data['hemo']=data['hemo'].fillna(data['hemo'].median())
data['pcv']=data['pcv'].fillna(data['pcv'].median())
```

```python
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
data['rbc']=l.fit_transform(data['rbc'])
data['pc']=l.fit_transform(data['pc'])
data['pcc']=l.fit_transform(data['pcc'])
data['ba']=l.fit_transform(data['ba'])
data['htn']=l.fit_transform(data['htn'])
data['dm']=l.fit_transform(data['dm'])
data['cad']=l.fit_transform(data['cad'])
data['appet']=l.fit_transform(data['appet'])
data['pe']=l.fit_transform(data['pe'])
data['ane']=l.fit_transform(data['ane'])
data['classification']=l.fit_transform(data['classification'])
data.head()

```

```python
x=data.drop(['classification'],axis=1)
y=data['classification']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)
```

```python
xtrain.shape,xtest.shape,ytrain.shape,ytest.shape
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(xtrain,ytrain)
```

```python
model.predict(xtest)
```

```python
model.score(xtest,ytest)
```
